```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from typing import List
```

### Helper Functions

```python
from typing import NamedTuple, Optional, Tuple, Generator

import numpy as np
from matplotlib import pyplot as plt
from skimage.draw import circle_perimeter_aa


class CircleParams(NamedTuple):
    row: int
    col: int
    radius: int


def draw_circle(img: np.ndarray, row: int, col: int, radius: int) -> np.ndarray:
    """
    Draw a circle in a numpy array, inplace.
    The center of the circle is at (row, col) and the radius is given by radius.
    The array is assumed to be square.
    Any pixels outside the array are ignored.
    Circle is white (1) on black (0) background, and is anti-aliased.
    """
    rr, cc, val = circle_perimeter_aa(row, col, radius)
    valid = (rr >= 0) & (rr < img.shape[0]) & (cc >= 0) & (cc < img.shape[1])
    img[rr[valid], cc[valid]] = val[valid]
    return img


def noisy_circle(
    img_size: int, min_radius: float, max_radius: float, noise_level: float
) -> Tuple[np.ndarray, CircleParams]:
    """
    Draw a circle in a numpy array, with normal noise.
    """

    # Create an empty image
    img = np.zeros((img_size, img_size))

    radius = np.random.randint(min_radius, max_radius)

    # x,y coordinates of the center of the circle
    row, col = np.random.randint(img_size, size=2)

    # Draw the circle inplace
    draw_circle(img, row, col, radius)

    added_noise = np.random.normal(0.5, noise_level, img.shape)
    img += added_noise

    return img, CircleParams(row, col, radius)


def show_circle(img: np.ndarray):
    fig, ax = plt.subplots()
    ax.imshow(img, cmap='gray')
    ax.set_title('Circle')
    plt.show()


def generate_examples(
    noise_level: float = 0.5,
    img_size: int = 100,
    min_radius: Optional[int] = None,
    max_radius: Optional[int] = None,
    dataset_path: str = 'ds',
) -> Generator[Tuple[np.ndarray, CircleParams], None, None]:
    if not min_radius:
        min_radius = img_size // 10
    if not max_radius:
        max_radius = img_size // 2
    assert max_radius > min_radius, "max_radius must be greater than min_radius"
    assert img_size > max_radius, "size should be greater than max_radius"
    assert noise_level >= 0, "noise should be non-negative"

    params = f"{noise_level=}, {img_size=}, {min_radius=}, {max_radius=}, {dataset_path=}"
    print(f"Using parameters: {params}")
    while True:
        img, params = noisy_circle(
            img_size=img_size, min_radius=min_radius, max_radius=max_radius, noise_level=noise_level
        )
        yield img, params


def iou(a: CircleParams, b: CircleParams) -> float:
    """Calculate the intersection over union of two circles"""
    r1, r2 = a.radius, b.radius
    d = np.linalg.norm(np.array([a.row, a.col]) - np.array([b.row, b.col]))
    if d > r1 + r2:
        return 0
    if d <= abs(r1 - r2):
        return 1
    r1_sq, r2_sq = r1**2, r2**2
    d1 = (r1_sq - r2_sq + d**2) / (2 * d)
    d2 = d - d1
    h1 = r1_sq * np.arccos(d1 / r1)
    h2 = d1 * np.sqrt(r1_sq - d1**2)
    h3 = r2_sq * np.arccos(d2 / r2)
    h4 = d2 * np.sqrt(r2_sq - d2**2)
    intersection = h1 + h2 + h3 + h4
    union = np.pi * (r1_sq + r2_sq) - intersection
    return intersection / union
```

### Dataset

```python
class CircleDataset(Dataset):
    def __init__(self, num_samples=1000):
        self.num_samples = num_samples
        self.generator = generate_examples()

    def __len__(self):
        return self.num_samples

    def __getitem__(self, idx):
        # We use the generator to provide items for our model
        img, circle_params = next(self.generator)

        # We return the image tensor and the circle params
        return torch.tensor(img, dtype=torch.float32).unsqueeze(0), torch.tensor(circle_params, dtype=torch.float32)
```

### Network

```python
class CNN(nn.Module):
    def __init__(self):
        super(CNN, self).__init__()
        # input images shape -> 1 * 100 * 100
        self.ConvBlock1 = nn.Sequential(
            nn.Conv2d(1, 32, 5), nn.BatchNorm2d(32), nn.ReLU(), nn.MaxPool2d(2, 2))
        # output shape -> 32 * 48 * 48
        self.ConvBlock2 = nn.Sequential(
            nn.Conv2d(32, 64, 3), nn.BatchNorm2d(64), nn.ReLU(), nn.MaxPool2d(2, 2))
        # output shape -> 64 * 23 * 23
        self.ConvBlock3 = nn.Sequential(
            nn.Conv2d(64, 128, 3), nn.BatchNorm2d(128), nn.ReLU(), nn.MaxPool2d(2, 2))
        # output shape -> 128 * 10 * 10
        self.ConvBlock4 = nn.Sequential(
            nn.Conv2d(128, 4, 1), nn.BatchNorm2d(4), nn.ReLU())
        # output shape -> 4 * 10 * 10
        self.FC1 = nn.Sequential(
            nn.Linear(4 * 10 * 10, 128), nn.ReLU(), nn.Linear(128, 16), nn.ReLU())
        self.FC2 = nn.Linear(16, 3)

    def forward(self, x):

        x = self.ConvBlock1(x)
        x = self.ConvBlock2(x)
        x = self.ConvBlock3(x)
        x = self.ConvBlock4(x)

        x = x.reshape(-1, 4*10*10)
        x = self.FC1(x)
        x = self.FC2(x)

        return x
```

### Training Loop

```python
def train_model(model, criterion, optimizer, trainloader, num_epochs=5):
    for epoch in range(num_epochs):
        running_loss = 0.0
        for inputs, targets in trainloader:

            inputs = inputs.to(device)
            targets = targets.to(device)

            optimizer.zero_grad()
            outputs = model(inputs)

            loss = criterion(outputs, targets/100) # scale our target to make sure they are between 0 and 1
            loss.backward()
            optimizer.step()
            running_loss += loss.item()

        print(f"Epoch {epoch + 1}, Loss: {running_loss / len(trainloader)}")
```

### Test Metric Functions

```python
def iou_value(pred_tensor: torch.Tensor, target_tensor: torch.Tensor) -> float:

    # The iou function is designed to accept tuples but our model gives us tensors
    # This function converts these tensors to the right format and makes calls to the provided iou function

    pred_tensor = pred_tensor.squeeze()
    target_tensor = target_tensor.squeeze()
    pred = CircleParams(float(pred_tensor[0]), float(
        pred_tensor[1]), float(pred_tensor[2]))
    target = CircleParams(float(target_tensor[0]), float(
        target_tensor[1]), float(target_tensor[2]))
    iou_value = iou(pred, target)

    return iou_value


def accuracy_over_iou_threshold(pred_iou_list: List[float], threshold: float = 0.5):

    # Calculate the fraction of test examples where our prediction did better than the iou threshold

    return sum([pred_iou >= threshold for pred_iou in pred_iou_list]) / len(pred_iou_list)
```

### Testing Loop

```python
def test_model(model, testloader, iou_threshold=0.75):
    model.eval()
    iou_list = [] # we keep track of iou scores over our test data to calculate the thresholded accuracy
    with torch.no_grad():
        for inputs, targets in testloader:
            inputs = inputs.to(device)
            targets = targets.to(device)
            outputs = model(inputs)
            iou = iou_value(outputs*100, targets) # rescale model outputs to match circle dimension of 100*100
            iou_list.append(iou)

        # Calculate accuracy for threshold
        acc = accuracy_over_iou_threshold(iou_list, iou_threshold)
        print(
            f"Test IoU Average Accuracy with IOU Threshold {iou_threshold}, Accuracy: {acc}")
```

### Model Training

Let's train our model. We generate a good amount of training samples and run our training loop. We save our model weights for reusability.

```python
train_dataset = CircleDataset(num_samples=100000)
trainloader = DataLoader(train_dataset, batch_size=16, shuffle=True)

device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")

model = CNN()
model.to(device)

criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
model.to(device)
```

```python
num_epochs = 30
train_model(model, criterion, optimizer, trainloader, num_epochs)

torch.save(model.state_dict(), 'circle_detection_cnn_wts.pth')
```

### Model Testing

Let's test our model by generating some test data and calculating the thresholed IOU score. We load the weights of our trained model and use it to get inferences on our test data. We take a look at accuracy scores for a few different threshold values.

```python
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
model = CNN()

if torch.cuda.is_available():
    # If GPU is available, load the model weights directly
    model.load_state_dict(torch.load('circle_detection_cnn_wts.pth'))
else:
    # If GPU is not available, load the model weights onto the CPU
    model.load_state_dict(torch.load('circle_detection_cnn_wts.pth', map_location=torch.device('cpu')))

model = model.to(device)
```

```python
test_dataset = CircleDataset(num_samples=200)
testloader = DataLoader(test_dataset, batch_size=1, shuffle=False)

thresholds = [0.5, 0.75, 0.9, 0.95]

for threshold in thresholds:
    test_model(model, testloader, threshold)
```

### Visualizing Results

Let's run a few sample images through our trained model to see how well it does. We'll show an image of the circle and take a look at the iou scores from our model prediction.

```python
for i in range(3):

    generator = generate_examples(0.5)
    x = next(generator)

    img_test = x[0]
    actual_params = x[1]

    show_circle(img_test)

    test_input = torch.tensor(
        img_test, dtype=torch.float32).unsqueeze(0).unsqueeze(0)
    test_input = test_input.to(device)
    prediction = model(test_input)

    print('Prediction: ', prediction*100)
    print('Actual Parameters: ', actual_params)
    print('IOU Value: ', iou_value(prediction*100,
          torch.tensor(actual_params, dtype=torch.float32)))
    print('\n')
```
